```python
#알고리즘 수업시간에 배운 동적프로그래밍
def fibo_with_dp(n):
    f[0] = 0
    f[1] = 1
    for i in range(2,n+1):
        f[i] = f[i-1] + f[i-2]
    return f[n]

n = int(input())
f = [0]*(n+2)
print(fibo_with_dp(n))
```

### 피보나치 함수 2020-07-22

```python
def fc_fibo(n):
    li = [[1,0],[0,1]]
    la = li.append
    for i in range(2,n+1):
        temp = []
        ta = temp.append
        for j,k in zip(li[i-1],li[i-2]):
            ta(j+k)
        la(temp)
    return li[n]
n = int(input())
for i in range(n):
    print(*fc_fibo(int(input())))
```

### RGB거리 2020-07-22

```python
#동적 프로그래밍-->이전에 계산된 값을 다시 이용할 수 있는가
#이전의 계산된 값을 이용하기 위해 반복문을 이용..!
n = int(input())
cost = [list(map(int, input().split())) for i in range(n)]

#(0:빨강 1:초록 2: 파랑)
#i번째 집을 빨강으로 칠하기는 초록, 파랑 이전 집의 최솟값에
#해당 비용을 더해주면 최솟값
for i in range(1,n):
    cost[i][0] = min(cost[i-1][1],cost[i-1][2]) + cost[i][0]
    cost[i][1] = min(cost[i-1][0],cost[i-1][2]) + cost[i][1]
    cost[i][2] = min(cost[i-1][0],cost[i-1][1]) + cost[i][2]
    
print(min(cost[n-1][0],cost[n-1][1],cost[n-1][2]))
```

### 정수 삼각형 2020-07-27

```python
def intTriangle(n, arr):
    #두번째 층부터 끝 층까지
    for i in range(1,n):
        #한 층의 원소를 모두 돈다.
        for j in range(i+1):
            #이전 층의 첫번째 원소일 때
            if j == 0:
                arr[i][j] = arr[i-1][0] + arr[i][j]
            #이전 층의 마지막 원소일 때
            elif j == i:
                arr[i][j] = arr[i-1][j-1] + arr[i][j]
            #그 외의 경우
            else:
                arr[i][j] = max(arr[i-1][j-1],arr[i-1][j]) + arr[i][j]
    return max(arr[n-1])

n = int(input())
arr = [list(map(int,input().split())) for i in range(n)]
print(intTriangle(n,arr))
```

### 계단 오르기 2020-07-27

```python
def climbing(n,arr):
    #시작점
    result[0] = 0
    #첫 번째 계단
    result[1] = arr[0]
    if n >= 2:
        #두 번째 계단
        result[2] = arr[0] + arr[1]
        for i in range(3,n+1):
            #마지막 전 계단을 밟았을 땐, 전전계단을 못밟음, 
            #마지막 전 계단을 안밟았을 때, 전전계단 까지의 최댓값
            result[i] = max(arr[i-1] + arr[i-2] + result[i-3], arr[i-1] + result[i-2])
    return result[n]

n = int(input())
result = [0] * (n+2)
arr = [int(input()) for i in range(n)]
print(climbing(n,arr))
```

### 쉬운 계단 수 2020-07-27

```python
# 갯수 세주는 함수 정의
def staircnt(n):
    #2차원 배열 정의 행은 1~9, 열은 n자리의 갯수 값
    result = [[0]*(n+1) for i in range(10)]
    #초깃값 정의
    for i in range(1,10):
        result[i][0] = 1
        if i != 9:
            result[i][1] = 2
        else:
            result[i][1] = 1
    #점화식 시작
    #1과 9일 때와 나머지일 때를 생각해야 함.
    for i in range(2,n):
        for j in range(1,10):
            if j == 1:
                result[j][i] = result[j][i-2] + result[j+1][i-1]
            elif j == 9:
                result[j][i] = result[j-1][i-1]
            else:
                result[j][i] = result[j-1][i-1] + result[j+1][i-1]
    total = 0
    #합산
    for i in range(1,10):
        total += result[i][n-1]
    return total

n = int(input())
print(staircnt(n) % 1000000000)
```

### 포도주 시식 2020-07-28

```python
#마지막 계단을 안밟을 수 있는 계단오르기 문제랑 똑같음.
def wine(n, arr):
    result = [0]*(n+1)
    result[1] = arr[0]
    if n >= 2:
        result[2] = arr[0] + arr[1]

        for i in range(3,n+1):
            result[i] = max(result[i-1],result[i-2] + arr[i-1], result[i-3] + arr[i-1] + arr[i-2])
    return result[n]
n = int(input())
arr = [int(input()) for i in range(n)]
print(wine(n,arr))
```

### 가장 긴 증가하는 부분 수열(LIS) 2020-07-28

```python
# 하 미치겠다 안풀려 계속틀려ㅣ아ㅜ피ㅏ두팹
# 7/31일 날 다시 풀기

def LIS(n, arr):
    result = [0] * n
    for i in range(n):
        for j in range(i):
            if arr[i] > arr[j] and result[i] < result[j]:
                result[i] = result[j]
        result[i] += 1
                
    return max(result)
            
    
n = int(input())
arr = list(map(int, input().split()))
print(LIS(n,arr))
```

### 가장 긴 감소하는 부분 수열(LDS) 2020-07-30

```python
def LDS(n,arr):
    result = [0]*n
    for i in range(n):
        for j in range(i):
            if arr[i] < arr[j] and result[i] < result[j]:
                result[i] = result[j]
        result[i]+=1
    return max(result)

n = int(input())
arr = list(map(int,input().split()))
print(LDS(n,arr))
```

### 가장 긴 바이토닉 부분 수열 2020-07-30

```python
#왼쪽에서 오른쪽으로의 LIS 배열과
#오른쪽에서 왼쪽으로의 LIS 배열의 합에서의 최댓값일까?

def bitonic(n,arr):
    LtoR = [0]*n
    RtoL = [0]*n
    inverse = arr[::-1]
    for i in range(n):
        for j in range(i):
            if arr[i] > arr[j] and LtoR[i] < LtoR[j]:
                LtoR[i] = LtoR[j]
            if inverse[i] > inverse[j] and RtoL[i]<RtoL[j]:
                RtoL[i] = RtoL[j]
        RtoL[i]+=1
        LtoR[i]+=1
    
    RtoL = RtoL[::-1]
    result = [i+j for i,j in zip(LtoR,RtoL)]
    return max(result) - 1

n = int(input())
arr = list(map(int, input().split()))
print(bitonic(n,arr))
```

### 가장 큰 증가 부분 수열 2020-07-30

```python
def BIS(n, arr):
    result = [0]*n
    for i in range(n):
        for j in range(i):
            if arr[i] > arr[j] and result[i] < result[j]:
                result[i] = result[j]
        result[i] += arr[i]
    
    return max(result)

n = int(input())
arr = list(map(int, input().split()))
print(BIS(n,arr))
```

### 전깃줄 2020-07-30

```python
#키 값 순서대로 정렬 한 뒤
#짝의 LIS를 찾아내면 됨..
#접근 : dictionary 자료형을 이용하고자 한다.

def electric(n, dic):
    #키 값 순서대로 정렬한 딕셔너리의 값들만 list에 담기
    li = [i[1] for i in sorted(dic.items())]
    #결과 배열
    result = [0] * n
    #LIS 알고리즘
    for i in range(n):
        for j in range(i):
            if li[i] > li[j] and result[i] < result[j]:
                result[i] = result[j]
        result[i]+=1
    #딕셔너리 길이에서 LIS를 빼면 빼야하는 전선의 최솟값.
    return len(dic) - max(result)

n = int(input())
arr = [list(map(int,input().split())) for i in range(n)]
dic = {i[0]:i[1] for i in arr}
print(electric(n,dic))
```

### LCS(Longest Common Subsequence) 2020-07-30

```python
def LCS(String1, String2):
    str1_len = len(String1)
    str2_len = len(String2)
    result = [[0]*(str1_len+1) for i in range(str2_len+1)]
    
    for i in range(1,str2_len+1):
        for j in range(1,str1_len+1):
            if String1[j-1] == String2[i-1]:
                result[i][j] = result[i-1][j-1]+1
            else:
                result[i][j] = max(result[i-1][j],result[i][j-1])
    return result[-1][-1]

String1 = input().upper()
String2 = input().upper()
print(LCS(String1,String2))
```
